```python
import pandas as pd
import io
import base64
import matplotlib.pyplot as plt
import seaborn as sns

# Load data safely
file_path = 'Students_Performance_data_set.csv'  # Make sure this file exists

try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
except Exception as e:
    print(f"Error loading DataFrame: {e}")
    df = pd.DataFrame()  # Empty DataFrame if loading fails

def plot_to_base64(fig):
    """Convert matplotlib figure to base64 string."""
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight')
    plt.close(fig)  # Properly close the figure
    buf.seek(0)
    return base64.b64encode(buf.read()).decode('utf-8')

if not df.empty:
    eda_markdown_output = "### 3. Exploratory Data Analysis (EDA)\n"
    eda_markdown_output += "#### Overview:\n"
    eda_markdown_output += ("This section explores the dataset to find patterns, distributions, and relationships "
                            "using statistical and visual analysis.\n\n")

    # Unique values in categorical columns
    categorical_cols = df.select_dtypes(include='object').columns
    eda_markdown_output += "#### Unique Values in Categorical Columns:\n"
    for col in categorical_cols:
        unique_vals = df[col].unique()
        eda_markdown_output += f"- **{col}:** {df[col].nunique()} unique values: {list(unique_vals)}\n"
    eda_markdown_output += "\n"

    # Histograms of scores
    fig1, axes1 = plt.subplots(1, 3, figsize=(18, 5))
    fig1.suptitle('Distribution of Student Scores', fontsize=16)

    sns.histplot(df['math score'], kde=True, ax=axes1[0], color='skyblue')
    axes1[0].set_title('Math Score Distribution')
    axes1[0].set_xlabel('Math Score')
    axes1[0].set_ylabel('Frequency')

    sns.histplot(df['reading score'], kde=True, ax=axes1[1], color='lightcoral')
    axes1[1].set_title('Reading Score Distribution')
    axes1[1].set_xlabel('Reading Score')
    axes1[1].set_ylabel('Frequency')

    sns.histplot(df['writing score'], kde=True, ax=axes1[2], color='lightgreen')
    axes1[2].set_title('Writing Score Distribution')
    axes1[2].set_xlabel('Writing Score')
    axes1[2].set_ylabel('Frequency')

    eda_markdown_output += "#### Visualizations: Distribution of Scores\n"
    eda_markdown_output += f"![Distribution of Scores](data:image/png;base64,{plot_to_base64(fig1)})\n\n"

    # Box plots by gender
    fig2, axes2 = plt.subplots(1, 3, figsize=(20, 6))
    fig2.suptitle('Scores by Gender', fontsize=16)

    sns.boxplot(x='gender', y='math score', data=df, ax=axes2[0], palette='pastel')
    axes2[0].set_title('Math Score by Gender')

    sns.boxplot(x='gender', y='reading score', data=df, ax=axes2[1], palette='pastel')
    axes2[1].set_title('Reading Score by Gender')

    sns.boxplot(x='gender', y='writing score', data=df, ax=axes2[2], palette='pastel')
    axes2[2].set_title('Writing Score by Gender')

    eda_markdown_output += "#### Visualizations: Scores by Gender\n"
    eda_markdown_output += f"![Scores by Gender](data:image/png;base64,{plot_to_base64(fig2)})\n\n"

    # Box plots by parental education
    education_order = [
        "some high school", "high school", "some college",
        "associate's degree", "bachelor's degree", "master's degree"
    ]

    fig3, axes3 = plt.subplots(1, 3, figsize=(22, 6))
    fig3.suptitle('Scores by Parental Level of Education', fontsize=16)

    sns.boxplot(x='parental level of education', y='math score', data=df,
                ax=axes3[0], palette='viridis', order=education_order)
    axes3[0].set_title('Math Score by Parental Education')
    axes3[0].tick_params(axis='x', rotation=45)

    sns.boxplot(x='parental level of education', y='reading score', data=df,
                ax=axes3[1], palette='viridis', order=education_order)
    axes3[1].set_title('Reading Score by Parental Education')
    axes3[1].tick_params(axis='x', rotation=45)

    sns.boxplot(x='parental level of education', y='writing score', data=df,
                ax=axes3[2], palette='viridis', order=education_order)
    axes3[2].set_title('Writing Score by Parental Education')
    axes3[2].tick_params(axis='x', rotation=45)

    fig3.tight_layout(rect=[0, 0.03, 1, 0.95])

    eda_markdown_output += "#### Visualizations: Scores by Parental Level of Education\n"
    eda_markdown_output += f"![Scores by Parental Education](data:image/png;base64,{plot_to_base64(fig3)})\n\n"

    # Correlation heatmap
    corr = df[['math score', 'reading score', 'writing score']].corr()
    fig4, ax4 = plt.subplots(figsize=(8, 6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=ax4)
    ax4.set_title('Correlation Heatmap of Scores')

    eda_markdown_output += "#### Visualizations: Correlation Heatmap of Scores\n"
    eda_markdown_output += f"![Correlation Heatmap](data:image/png;base64,{plot_to_base64(fig4)})\n"

    print(eda_markdown_output)

else:
    print("DataFrame is empty. Please check the dataset loading step.")
```

```python
import pandas as pd

# Load dataset if not already loaded
try:
    if 'df' not in globals():
        file_path = 'Students_Performance_data_set.csv'  # Adjust filename if needed
        df = pd.read_csv(file_path)
except Exception as e:
    print(f"Error loading dataset: {e}")
    df = pd.DataFrame()  # Empty DataFrame to avoid errors below

# Prepare the cleaning report
cleaning_report = "### 4. Data Cleaning & Preprocessing\n"
cleaning_report += "#### Overview:\n"
cleaning_report += "This section checks for missing values and duplicates to ensure dataset quality.\n\n"

if not df.empty:
    # Missing values
    missing_values = df.isnull().sum()
    cleaning_report += "**Missing Values:**\n"
    cleaning_report += missing_values.to_markdown() + "\n\n"

    # Duplicate rows
    duplicate_count = df.duplicated().sum()
    cleaning_report += f"**Duplicate Rows:** {duplicate_count}\n"

    if duplicate_count > 0:
        df.drop_duplicates(inplace=True)
        cleaning_report += "Duplicate rows removed.\n\n"
    else:
        cleaning_report += "No duplicate rows found.\n\n"

    cleaning_report += "**Summary of Cleaning:**\n"
    cleaning_report += ("No missing values and no duplicates detected. "
                        "No normalization or standardization applied since not required for analysis. "
                        "Categorical encoding will be applied as needed during analysis.\n")
else:
    cleaning_report += "Dataset is empty. No cleaning performed.\n"

print(cleaning_report)
```

```python
import pandas as pd
from scipy import stats

statistical_markdown_output = ""

# 1. T-test: Impact of 'test preparation course' on 'math score'
completed_course = df[df['test preparation course'] == 'completed']['math score']
not_completed_course = df[df['test preparation course'] == 'none']['math score']

if not completed_course.empty and not not_completed_course.empty:
    t_stat, p_val = stats.ttest_ind(completed_course, not_completed_course, equal_var=False)
    statistical_markdown_output += (
        "**1. T-test for Math Score by Test Preparation Course:**\n"
        f"- Completed group mean: {completed_course.mean():.2f}\n"
        f"- None group mean: {not_completed_course.mean():.2f}\n"
        f"- T-statistic: {t_stat:.3f}\n"
        f"- P-value: {p_val:.3f}\n"
    )
    if p_val < 0.05:
        statistical_markdown_output += "- Conclusion: Reject null hypothesis. Significant difference exists.\n\n"
    else:
        statistical_markdown_output += "- Conclusion: Fail to reject null hypothesis. No significant difference.\n\n"
else:
    statistical_markdown_output += "- Not enough data to perform T-test for math scores by test preparation course.\n\n"

# 2. ANOVA: Impact of 'parental level of education' on 'reading score'
education_levels = [
    "some high school", "high school", "some college",
    "associate's degree", "bachelor's degree", "master's degree"
]

groups = [df.loc[df['parental level of education'] == level, 'reading score'].dropna() for level in education_levels]
groups = [g for g in groups if not g.empty]

if len(groups) > 1:
    f_stat, p_val = stats.f_oneway(*groups)
    statistical_markdown_output += (
        "**2. ANOVA for Reading Score by Parental Level of Education:**\n"
        f"- F-statistic: {f_stat:.3f}\n"
        f"- P-value: {p_val:.3f}\n"
    )
    if p_val < 0.05:
        statistical_markdown_output += "- Conclusion: Reject null hypothesis. Significant differences between groups.\n\n"
    else:
        statistical_markdown_output += "- Conclusion: Fail to reject null hypothesis. No significant differences between groups.\n\n"
else:
    statistical_markdown_output += "- Not enough groups with data for ANOVA on reading scores.\n\n"

# 3. Chi-squared test: Relationship between 'gender' and 'test preparation course'
contingency = pd.crosstab(df['gender'], df['test preparation course'])
chi2, p_val, dof, expected = stats.chi2_contingency(contingency)

statistical_markdown_output += (
    "**3. Chi-squared Test for Gender vs Test Preparation Course:**\n"
    f"- Contingency Table:\n{contingency.to_markdown()}\n"
    f"- Chi-squared statistic: {chi2:.3f}\n"
    f"- P-value: {p_val:.3f}\n"
)

if p_val < 0.05:
    statistical_markdown_output += "- Conclusion: Reject null hypothesis. There is a significant association.\n\n"
else:
    statistical_markdown_output += "- Conclusion: Fail to reject null hypothesis. No significant association.\n\n"

# Interpretation
statistical_markdown_output += (
    "**Interpretation of Findings:**\n"
    "- Math scores are significantly higher for students who completed the test preparation course.\n"
    "- Reading scores vary significantly with parental education level.\n"
    "- Gender and test preparation course completion are significantly associated.\n"
    "- Overall, these findings suggest test preparation and parental education influence student performance, and gender relates to course completion behavior.\n"
)

print(statistical_markdown_output)
```
